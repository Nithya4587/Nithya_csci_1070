```python
jeopardy_df = pd.read_csv('JEOPARDY_DATA.csv')
jeopardy_df.head()
```

```python
jeopardy_df.columns = jeopardy_df.columns.str.replace(" ","")
jeopardy_df.columns
```

```python
jeopardy_df_round = pd.get_dummies(jeopardy_df[['Round']])
jeopardy_df_round
```

```python
jeopardy_df.dtypes
```

```python
jeopardy_df['Value']=jeopardy_df['Value'].str.replace("$","").str.replace(',','').str.replace(' ','')
jeopardy_df
```

```python
jeopardy_df['Value'] = jeopardy_df['Value'].replace("None", value=np.nan)
jeopardy_df['Value']=jeopardy_df['Value'].astype(float)
```
