```python
import pandas as pd
import numpy as np
```

```python
survey_df = pd.read_csv('CSCI_1070_survey_results.csv')
survey_df.head(20)
```

```python
survey_df.rename(columns = {"What state/country were you born in?":"Birthplace"}, inplace = True)
survey_df.head(20)
```

```python
Birthplace_map = {'Missouri':1, 'Saint Louis, Missouri':1, 'Illinois':2, 'Japan':3, 'United States':4, 'Saudi Arabia':5, 'Minnesota':6, 'India':7, 'New Mexico':8, 'United States, Missouri':1, 'Missouri/U.S':1, 'Georgia, USA':9, 'Benghazi, Libya':10}
survey_df['Birthplace'] = survey_df['Birthplace'].map(Birthplace_map)
survey_df
```

```python
survey_df.rename(columns = {"How many pets and of which kind do you have?":"Number of Pets"}, inplace = True)
survey_df.head(20)
```

```python
survey_df.rename(columns = {"What is your favorite type of food?":"Favorite Food"}, inplace = True)
survey_df.head(20)
```

```python
survey_df.rename(columns = {"What is your major?":"Major"}, inplace = True)
survey_df.head(20)
```

```python
survey_df.rename(columns = {"How old are you?":"Age"}, inplace = True)
survey_df.head(20)
```

```python
Age_map = {'18':18, '18 years old':18, '19':19, '19 turning 20 on February 28':19, '20':20, '21':21, '22':22}
survey_df['Age'] = survey_df['Age'].map(Age_map)
survey_df
```

```python
import matplotlib.pyplot as plt
```

```python
survey_df['Do you take the metro?'].value_counts()
```

```python
#This histograph represents how many students take the metro and how many don't.
plt.figure(figsize = [3,5])
plt.hist(survey_df['Do you take the metro?'])
plt.xlabel("Take the Metro")
plt.ylabel("Number of People")
plt.title("Students That Take the Metro")
```

```python
agg_age = survey_df.groupby('Age').count()
agg_age
```

```python
#This pie chart shows the ratio of the different ages of the students in the class.
#This showcases that there is a higher percentage of 18 year olds in the class
colors = ['green','blue','yellow','pink','red']
agg_age.plot(kind='pie',y='Timestamp',autopct = '%1.0f%%', colors=colors)
```

```python
agg_work = survey_df.groupby('Are you working while in school?').count()
agg_work
```

```python
# This pie chart shows the ratio of how many students work in the class to how many don't.
colors = ['purple','red']
agg_work.plot(kind='pie',y='Timestamp',autopct = '%1.0f%%', colors=colors)
```

```python
agg_sibling = survey_df.groupby('How many siblings do you have?').count()
agg_sibling
```

```python
# This scatter plot shows the differences in where people were born and compares them.
plt.plot(survey_df['Birthplace'],'.')
plt.xlabel('Number of Person')
plt.ylabel('Birthplace')
```

```python
#This histogram shows the how many people have how many siblings.
n, bins, patches = plt.hist(x=survey_df['How many siblings do you have?'],bins='auto')
plt.xlabel('Number of Siblings')
plt.ylabel('Number of people')
```
